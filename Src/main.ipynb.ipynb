```python
from datasets import load_dataset
import pandas as pd

# Load datasets
medmcqa = load_dataset("openlifescienceai/medmcqa", split="train")
medquad = load_dataset("lavita/MedQuAD", split="train")
pubmedqa = load_dataset("qiaojin/PubMedQA", "pqa_labeled", split="train")  # Fixed config

# Keywords for filtering
cancer_keywords = ["cancer", "breast", "lung", "oral", "cervical", "tumor", "oncology", "neoplasm"]

def filter_cancer_rows(df, question_col, answer_col):
    return df[
        df[question_col].str.contains("|".join(cancer_keywords), case=False, na=False) |
        df[answer_col].str.contains("|".join(cancer_keywords), case=False, na=False)
    ]

# Convert datasets to DataFrames and rename columns
df_medmcqa = medmcqa.to_pandas().rename(columns={"question": "question", "exp": "answer"})
df_medquad = medquad.to_pandas().rename(columns={"Question": "question", "Answer": "answer"})
df_pubmedqa = pubmedqa.to_pandas().rename(columns={"question": "question", "final_decision": "answer"})

# Add source column
df_medmcqa["source"] = "MedMCQA"
df_medquad["source"] = "MedQuAD"
df_pubmedqa["source"] = "PubMedQA"

# Filter rows
df_medmcqa_filtered = filter_cancer_rows(df_medmcqa, "question", "answer")
df_medquad_filtered = filter_cancer_rows(df_medquad, "question", "answer")
df_pubmedqa_filtered = filter_cancer_rows(df_pubmedqa, "question", "answer")

# Dynamic sampling: take min(rows, desired_sample)
def safe_sample(df, n):
    return df.sample(n=min(n, len(df)), random_state=42)

df_medmcqa_filtered = safe_sample(df_medmcqa_filtered, 2500)
df_medquad_filtered = safe_sample(df_medquad_filtered, 2500)
df_pubmedqa_filtered = safe_sample(df_pubmedqa_filtered, 2500)

# Combine datasets
final_df = pd.concat([df_medmcqa_filtered, df_medquad_filtered, df_pubmedqa_filtered])
final_df = final_df.sample(n=min(10000, len(final_df)), random_state=42).reset_index(drop=True)

# Save to CSV
final_df.to_csv("cancer_qa_finetune_10k.csv", index=False)
print(" Saved cancer_qa_finetune_10k.csv with shape:", final_df.shape)
```

```python
import pandas as pd
from datasets import Dataset

# Load your CSV into a Hugging Face Dataset
dataset = Dataset.from_pandas(pd.read_csv("cancer_qa_finetune_10k.csv"))

# Shuffle the dataset
shuffled = dataset.shuffle(seed=42)

# Select all rows (since you only have ~5,131)
training_dataset = shuffled.select(range(len(shuffled)))

print(training_dataset)
```

```python
import bitsandbytes as bnb
from transformers import BitsAndBytesConfig

quantization_config=BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype="float16")
```

```python
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name="BioMistral/BioMistral-7B"
model=AutoModelForCausalLM.from_pretrained(model_name,quantization_config=quantization_config,device_map="auto")
model.config.use_cache=True
tokenizer=AutoTokenizer.from_pretrained(model_name,trust_remote_code=True)
# Set the pad token to be different from the eos token
tokenizer.pad_token = tokenizer.unk_token
tokenizer.padding_side="right"
```

```python
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from transformers import GenerationConfig

# LoRA Config
peft_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=['q_proj', 'k_proj', 'down_proj', 'v_proj', 'gate_proj', 'o_proj', 'up_proj'],
    lora_dropout=0.05,
    task_type="CAUSAL_LM"
)

# Prepare model for QLoRA
model = prepare_model_for_kbit_training(model)
model = get_peft_model(model, peft_config)
model.gradient_checkpointing_enable()

# Print trainable parameters
def print_trainable_parameters(model):
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    total = sum(p.numel() for p in model.parameters())
    print(f"Trainable params: {trainable} / {total} ({100 * trainable / total:.2f}%)")

print_trainable_parameters(model)

# Generation Config
generation_config = GenerationConfig(
    pad_token_id=tokenizer.eos_token_id,
    eos_token_id=tokenizer.eos_token_id,
    max_new_tokens=256,
    temperature=0.7,
    top_p=0.9,
    do_sample=True
)
model.generation_config = generation_config
```

```python
import torch

def generate(prompt):
    max_new_tokens = 50

    encoded = tokenizer(prompt,return_tensors="pt",padding=True,truncation=True,max_length=512).to(device)

    with torch.inference_mode():
        out = model.generate(
            input_ids=encoded["input_ids"],
            attention_mask=encoded["attention_mask"],
            max_new_tokens=max_new_tokens,
            temperature=0.7,
            top_p=0.9,
            repetition_penalty=1.2,
            do_sample=True
        )

    # Remove the input tokens from the output
    generated_tokens = out[0][encoded["input_ids"].shape[-1]:]
    response = tokenizer.decode(generated_tokens, skip_special_tokens=True)
    print(response.strip())
    return response.strip()

# Test
generate("Explain the role of mitochondria in cells and tissues.")
```

```python
# Re-tokenize dataset with labels
def tokenize_function(examples):
    texts = [
        f"Question: {q}\nAnswer: {a}"
        for q, a in zip(examples["question"], examples["answer"])
    ]
    model_inputs = tokenizer(
        texts,
        truncation=True,
        padding="max_length",
        max_length=512
    )
    # Set labels same as input_ids for CausalLM
    model_inputs["labels"] = model_inputs["input_ids"].copy()
    return model_inputs

# Apply tokenization
tokenized_dataset = training_dataset.map(
    tokenize_function,
    batched=True,
    remove_columns=['question', 'answer']
)
```

```python
from transformers import DataCollatorForLanguageModeling

# Create the data collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False   # We're not doing masked language modeling
)
```

```python
# Training arguments (no evaluation)
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="fine_tuning_BioMistral",
    per_device_train_batch_size=2,      # Adjust if VRAM is low
    gradient_accumulation_steps=8,      # Simulate larger batch
    num_train_epochs=1,                 # Train for 1 epoch first
    learning_rate=5e-4,
    weight_decay=0.01,
    fp16=True,
    optim="paged_adamw_8bit",
    lr_scheduler_type="cosine",
    warmup_ratio=0.1,
    save_strategy="steps",
    save_steps=50,                      # Save at the end of each epoch
    logging_steps=50,
    report_to="none"                    # Disable wandb/other loggers
)

# Trainer Setup (no eval_dataset)
from transformers import Trainer

trainer = Trainer(
    model=model,
    train_dataset=tokenized_dataset,
    args=training_args,
    data_collator=data_collator
)

# Disable cache for training
model.config.use_cache = False
```

```python
# Train
trainer.train()
```

```python
import math
from datasets import Dataset

# Load a small test set (or split some data before training)
test_dataset = tokenized_dataset.select(range(500))  # First 500 samples

# Evaluate loss
eval_results = trainer.evaluate(test_dataset)
perplexity = math.exp(eval_results["eval_loss"])
print(f"Perplexity: {perplexity}")
```

```python
from llama_index.readers.papers.pubmed.base import PubmedReader
import os
import json
import time
import sys
import contextlib

# Initialize reader
pub_reader = PubmedReader()

# Define cancer types
#cancer_types = ["lung cancer", "breast cancer", "oral cancer", "cervical cancer"]
cancer_types=["cervical cancer"]
# Google Drive path
BASE_PATH = "/content/drive/MyDrive/majorproject/data/pubmed"
os.makedirs(BASE_PATH, exist_ok=True)

# Limit per cancer type
MAX_RESULTS = 1250

# Suppress unwanted prints from PubmedReader
@contextlib.contextmanager
def suppress_stdout():
    with open(os.devnull, "w") as devnull:
        old_stdout = sys.stdout
        sys.stdout = devnull
        try:
            yield
        finally:
            sys.stdout = old_stdout

def fetch_and_save_pubmed(cancer_query, max_results=MAX_RESULTS):
    print(f" Fetching papers for: {cancer_query}")

    # Silence internal PubmedReader output
    with suppress_stdout():
        docs = pub_reader.load_data(search_query=cancer_query, max_results=max_results)

    folder_name = cancer_query.replace(" ", "_")
    folder_path = os.path.join(BASE_PATH, folder_name)
    os.makedirs(folder_path, exist_ok=True)

    for i, doc in enumerate(docs):
        paper = {
            "title": doc.metadata.get("title", f"Paper_{i+1}"),
            "abstract": doc.text,
            "metadata": doc.metadata
        }
        file_path = os.path.join(folder_path, f"paper_{i+1}.json")
        with open(file_path, "w", encoding="utf-8") as f:
            json.dump(paper, f, ensure_ascii=False, indent=2)

    print(f" Successfully stored {len(docs)} papers in folder: {folder_name}\n")
    time.sleep(1)

# Fetch for all cancer types
for cancer in cancer_types:
    fetch_and_save_pubmed(cancer)
```

```python
from tqdm import tqdm
import re

def split_abstract(abstract, max_chars=1200):
    """Chunk abstracts by paragraph/section, else every N chars."""
    # Try to split by double newline or section/numbered headings if available
    # If not, fallback to every ~1200 chars
    chunks = []
    if "\n\n" in abstract:
        parts = [p.strip() for p in abstract.split('\n\n') if p.strip()]
    else:
        # fallback: split every max_chars with overlap
        parts = [abstract[i:i+max_chars] for i in range(0, len(abstract), max_chars)]
    # Clean and filter out too-short chunks
    for p in parts:
        if len(p) > 120:
            chunks.append(p)
    return chunks

chunked_papers = []
total_chunks = 0
chunk_sizes = []

for cancer_type, file_path in tqdm(all_files, desc="Chunking papers"):
    with open(file_path, "r") as f:
        data = json.load(f)
    title = data.get('title', '')
    abstract = data.get('abstract', '')
    metadata = data.get('metadata', {})
    # Normalize cancer type (underscore to space)
    cancertype_clean = cancer_type.replace("_", " ")

    # Split abstract into chunks (by paragraph or every 1200 chars)
    chunks = split_abstract(abstract, max_chars=1200)
    for idx, chunk in enumerate(chunks):
        chunk_data = {
            "title": title,
            "chunk_text": chunk,
            "metadata": {
                **metadata,
                "cancer_type": cancertype_clean
            },
            "chunk_id": idx + 1
        }
        chunk_sizes.append(len(chunk))
        chunked_papers.append(chunk_data)
    total_chunks += len(chunks)

# Calculation summaries
print(f"\nTotal papers processed: {len(all_files)}")
print(f"Total chunks created: {total_chunks}")
print(f"Average chunk length (chars): {int(sum(chunk_sizes)/len(chunk_sizes))}")
print(f"Metadata fields in sample: {list(chunked_papers[0]['metadata'].keys())}")
print(f"Sample chunk preview:\n{chunked_papers[0]}")
```

```python
import PyPDF2
import docx
from PIL import Image
import pytesseract
import re
import spacy

# Load medical NLP model (install: python -m spacy download en_core_web_sm)
nlp = spacy.load("en_core_web_sm")

class DocumentProcessor:
    def __init__(self):
        self.cancer_types = ["breast cancer", "lung cancer", "cervical cancer", "oral cancer"]
        self.biomarkers = ["HER2", "ER", "PR", "EGFR", "KRAS", "BRCA1", "BRCA2", "PD-L1"]

    def extract_text_from_pdf(self, pdf_file):
        """Extract text from PDF file"""
        text = ""
        reader = PyPDF2.PdfReader(pdf_file)
        for page in reader.pages:
            text += page.extract_text()
        return text

    def extract_text_from_docx(self, docx_file):
        """Extract text from DOCX file"""
        doc = docx.Document(docx_file)
        text = ""
        for paragraph in doc.paragraphs:
            text += paragraph.text + "\n"
        return text

    def extract_text_from_image(self, image_file):
        """Extract text from image using OCR"""
        image = Image.open(image_file)
        text = pytesseract.image_to_string(image)
        return text

    def extract_medical_entities(self, text):
        """Extract structured medical information from text"""
        doc = nlp(text.lower())

        extracted_info = {
            "cancer_type": None,
            "stage": None,
            "biomarkers": [],
            "treatments": [],
            "raw_text": text
        }

        # Extract cancer type
        for cancer in self.cancer_types:
            if cancer in text.lower():
                extracted_info["cancer_type"] = cancer.replace(" ", "_")
                break

        # Extract biomarkers
        for marker in self.biomarkers:
            pattern = rf"{marker}[+-]|\b{marker}\b"
            if re.search(pattern, text, re.IGNORECASE):
                extracted_info["biomarkers"].append(marker)

        # Extract stage information
        stage_pattern = r"stage\s*([IV]{1,3}|[1-4]|[ABCD])"
        stage_match = re.search(stage_pattern, text, re.IGNORECASE)
        if stage_match:
            extracted_info["stage"] = stage_match.group(1)

        return extracted_info

# Initialize processor
doc_processor = DocumentProcessor()
```

```python
class QueryProcessor:
    def __init__(self):
        self.intent_keywords = {
            "drug_recommendation": ["drug", "treatment", "therapy", "medication", "recommend"],
            "side_effects": ["side effect", "adverse", "reaction", "toxicity"],
            "prognosis": ["prognosis", "survival", "outcome", "life expectancy"],
            "comparison": ["vs", "versus", "compare", "better", "difference"],
            "clinical_trial": ["trial", "study", "research", "experimental"]
        }

    def classify_intent(self, query):
        """Classify user query intent"""
        query_lower = query.lower()
        intents = []

        for intent, keywords in self.intent_keywords.items():
            if any(keyword in query_lower for keyword in keywords):
                intents.append(intent)

        return intents if intents else ["general_query"]

    def extract_query_entities(self, query):
        """Extract medical entities from user query"""
        doc = nlp(query.lower())

        entities = {
            "cancer_type": None,
            "drugs": [],
            "biomarkers": [],
            "intent": self.classify_intent(query)
        }

        # Extract cancer type from query
        for cancer in doc_processor.cancer_types:
            if cancer in query.lower():
                entities["cancer_type"] = cancer.replace(" ", "_")
                break

        # Extract drug names (simple pattern matching)
        drug_patterns = [
            r"trastuzumab", r"herceptin", r"pembrolizumab", r"keytruda",
            r"bevacizumab", r"avastin", r"erlotinib", r"tarceva",
            r"gefitinib", r"iressa", r"osimertinib", r"tagrisso"
        ]

        for pattern in drug_patterns:
            if re.search(pattern, query, re.IGNORECASE):
                entities["drugs"].append(pattern)

        return entities

# Initialize query processor
query_processor = QueryProcessor()
```

```python
class MedicalQAInput:
    def __init__(self):
        self.doc_processor = DocumentProcessor()
        self.query_processor = QueryProcessor()
        self.session_context = {}

    def process_document_upload(self, file, file_type):
        """Process uploaded medical document"""
        try:
            # Extract text based on file type
            if file_type == "pdf":
                text = self.doc_processor.extract_text_from_pdf(file)
            elif file_type == "docx":
                text = self.doc_processor.extract_text_from_docx(file)
            elif file_type in ["jpg", "png", "jpeg"]:
                text = self.doc_processor.extract_text_from_image(file)
            else:
                return {"error": "Unsupported file type"}

            # Extract medical entities
            extracted_info = self.doc_processor.extract_medical_entities(text)

            # Store in session context
            self.session_context.update(extracted_info)

            # Generate suggested questions
            suggested_questions = self.generate_suggested_questions(extracted_info)

            return {
                "status": "success",
                "extracted_info": extracted_info,
                "suggested_questions": suggested_questions,
                "message": f"Document processed successfully. Found: {extracted_info['cancer_type'] or 'Unknown cancer type'}"
            }

        except Exception as e:
            return {"error": f"Failed to process document: {str(e)}"}

    def process_text_query(self, query):
        """Process text-based user query"""
        # Extract entities from query
        query_entities = self.query_processor.extract_query_entities(query)

        # Merge with session context (if document was uploaded)
        combined_context = {**self.session_context, **query_entities}

        # Clean and prepare query for downstream processing
        processed_query = {
            "original_query": query,
            "entities": combined_context,
            "cancer_type": combined_context.get("cancer_type"),
            "intent": query_entities["intent"],
            "has_document_context": bool(self.session_context)
        }

        return processed_query

    def generate_suggested_questions(self, extracted_info):
        """Generate context-aware suggested questions"""
        suggestions = []
        cancer_type = extracted_info.get("cancer_type", "").replace("_", " ")

        if cancer_type:
            suggestions.extend([
                f"What are the latest treatment options for {cancer_type}?",
                f"What are the side effects of standard {cancer_type} treatments?",
                f"Are there any clinical trials available for {cancer_type}?"
            ])

        if extracted_info.get("biomarkers"):
            markers = ", ".join(extracted_info["biomarkers"])
            suggestions.append(f"What targeted therapies are available for {markers} positive tumors?")

        # Generic suggestions
        suggestions.extend([
            "What is my prognosis?",
            "What should I ask my doctor?",
            "Are there alternative treatment options?"
        ])

        return suggestions[:5]  # Return top 5 suggestions

    def reset_session(self):
        """Reset session context"""
        self.session_context = {}

# Initialize main input handler
medical_input = MedicalQAInput()
```

```python
import re

def clean_answer(answer, drug_names):
    # De-duplicate lines and standardize spacing
    lines = [line.strip() for line in answer.split('\n') if line.strip()]
    unique = []
    for line in lines:
        if line not in unique:
            unique.append(line)
    answer = '\n'.join(unique)

    # Highlight drug names (bold)
    for drug in drug_names:
        answer = re.sub(rf"\b{re.escape(drug)}\b", f"**{drug}**", answer, flags=re.IGNORECASE)

    # Bullet treatments, split by clinical context
    bullet_lines = []
    for line in answer.split('\n'):
        if "treatment" in line.lower() or "therapy" in line.lower():
            bullet_lines.append(f"- {line}")
        else:
            bullet_lines.append(line)
    answer = '\n'.join(bullet_lines)

    return answer
```

```python
def format_sources(chunks):
    sources = set()
    for chunk in chunks:
        # Extract 1st sentence or title or any reference text as preview
        preview = chunk.strip().split('\n')[0]
        preview = preview[:120] + ('...' if len(preview) > 120 else '')
        sources.add(preview)
    if not sources:
        return "No matching documents found—answer generated by LLM only."
    return "\n".join(f"- {src}" for src in sources)
```

```python
def plausibility_check(answer, chunks):
    if not chunks:
        return "⚠️ No matching research found. This is a general AI-generated answer. Ask for more details or upload your medical report."
    if "uncertain" in answer.lower() or "unknown" in answer.lower():
        return "⚠️ The information may be limited or unclear based on current research. Please consult your oncologist."
    return ""
```

```python
def suggest_follow_ups(drugs, cancer_type):
    suggestions = []
    if drugs:
        drug = drugs[0]
        suggestions.append(f"What are the side effects of {drug}?")
        suggestions.append(f"Is {drug} covered in recent clinical trials for {cancer_type.replace('_', ' ')}?")
    if cancer_type:
        suggestions.append(f"Are there alternative therapies for {cancer_type.replace('_', ' ')}?")
        suggestions.append("What is the prognosis with current treatments?")
    suggestions.append("How do these treatments compare?")
    return suggestions[:4]
```

```python
def format_final_display(user_query, answer, chunks, drugs, cancer_type):
    cleaned = clean_answer(answer, drugs)
    sources = format_sources(chunks)
    plausibility = plausibility_check(answer, chunks)
    followups = suggest_follow_ups(drugs, cancer_type)

    display = f"""
**Patient Question:**
{user_query}

**Clinical Answer:**
{cleaned}

{plausibility}

**Cited Research Chunks:**
{sources}

**Suggested Follow-up Questions:**
""" + "\n".join(f"- {q}" for q in followups)
    return display

# Example Usage:
user_query = "What are the best treatments for HER2 positive breast cancer?"
drug_names = ["Trastuzumab", "Pertuzumab", "DS-8201", "Paclitaxel", "Alectinib", "Durvalumab"]
cancer_type = "breast_cancer"

# From your previous pipeline
# answer = <BioMistral LLM answer text>
# retrieved_chunks = <list of top BM25 chunks>

full_response = format_final_display(user_query, answer, retrieved_chunks, drug_names, cancer_type)
print(full_response)
```

# Final Evaluation Metrics

```python
# Recall@k: Percentage of gold answers found in top-k retrieved chunks
def recall_at_k(retrieved_chunks, gold_answers, k=5):
    recall_count = 0
    for gold in gold_answers:
        found = any(gold.lower() in chunk.lower() for chunk in retrieved_chunks[:k])
        if found:
            recall_count += 1
    recall = recall_count / len(gold_answers) if gold_answers else 0
    return recall

# Precision@k: Fraction of top-k chunks that are relevant to any gold answer
def precision_at_k(retrieved_chunks, gold_answers, k=5):
    relevant = 0
    for chunk in retrieved_chunks[:k]:
        if any(gold.lower() in chunk.lower() for gold in gold_answers):
            relevant += 1
    precision = relevant / k
    return precision

# MRR: Reciprocal rank of the first relevant retrieved chunk
def mrr(retrieved_chunks, gold_answers, k=5):
    for rank, chunk in enumerate(retrieved_chunks[:k], start=1):
        if any(gold.lower() in chunk.lower() for gold in gold_answers):
            return 1.0 / rank
    return 0.0

# Coverage: Fraction of expected drugs/treatments mentioned anywhere in the retrieved context
def coverage(retrieved_chunks, expected_terms):
    retrieved_text = " ".join(retrieved_chunks).lower()
    matched = sum(1 for term in expected_terms if term.lower() in retrieved_text)
    return matched / len(expected_terms) if expected_terms else 0
```

```python
def factuality_score(answer, approved_drugs):
    # Returns True if any valid drug is found, False otherwise
    return any(drug.lower() in answer.lower() for drug in approved_drugs)

def hallucination_rate(generated_answers, approved_drugs):
    def has_hallucination(answer):
        answer_drugs = [w for w in answer.split() if w.lower() not in [d.lower() for d in approved_drugs]]
        # You can use a smarter NER/extraction if needed
        return any(word for word in answer_drugs if word.istitle() and word not in approved_drugs)
    hallucinations = [ans for ans in generated_answers if not factuality_score(ans, approved_drugs)]
    return len(hallucinations) / len(generated_answers) if generated_answers else 0

def coverage_score(answer, expected_terms):
    return sum(1 for term in expected_terms if term.lower() in answer.lower()) / len(expected_terms) if expected_terms else 0


import textstat

def readability(answer):
    # Flesch Reading Ease: higher is easier (60–90 = easy patient text)
    return textstat.flesch_reading_ease(answer)

```
